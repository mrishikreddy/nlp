1.	Implement Word Embeddings on following text:
    
          words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england'].


```python
import gensim.downloader as api
model=api.load("glove-wiki-gigaword-50")
words=['king','queen','man','woman','paris','france','london','england']
for word in words:
    if word in model:
        vector=model[word]
        print(f"Word: {word}\nVector: {vector[:5]}....\n")
    else:
        print(f"Word:{word} not in vocabulary.\n")
```

2.	Implement Word Embeddings using PCA (Principal Component Analysis) on following text:

          words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england'].


```python
import gensim.downloader as api
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
model=api.load("glove-wiki-gigaword-50")
words=['king','queen','man','woman','paris','france','london','england']
embeddings=[model[word] for word in words]
pca=PCA(n_components=2)
reduced_embeddings=pca.fit_transform(embeddings)
plt.figure(figsize=(8,6))
for i,word in enumerate(words):
    plt.scatter(reduced_embeddings[i, 0], reduced_embeddings[i, 1])
    plt.text(reduced_embeddings[i, 0] + 0.01, reduced_embeddings[i, 1] + 0.01, word, fontsize=12)

plt.title("PCS of Word Embeddings")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```

3.	Execute Dependency Parsing on any text or on following text :
          The quick brown fox jumps over the lazy dog


```python
import spacy
nlp=spacy.load("en_core_web_sm")
sentence="The quick brown fox jumps over the lazy dog."
doc=nlp(sentence)
for token in doc:
    print(f"Word:{token.text}, POS:{token.pos_}, Dependency:{token.dep_}, Head:{token.head.text}")
```
