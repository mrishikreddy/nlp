```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Alice', None],
    'Age': [25, None, 23, 35, None, 45, 25, 29],
    'Gender': ['F', 'M', 'M', 'M', None, 'M', 'F', 'f'],
    'City': ['New York', 'Los Angeles', 'New York', 'Houston', 'Los Angeles', 'Houston', 'New York', 'New York']
}
df = pd.DataFrame(data)
df.head(10)
```

```python
print("the number of missing values n each column are:")
print(df.isnull().sum())
```

```python
df_cleaned = df.dropna()
print("the cleaned data sets with different ways are:")
print(df_cleaned)
```

Note : The df and df_cleaned will hold same values because of shallow copy

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Name'] = df['Name'].fillna('Unknown')
print(df)
```

```python
print("Number of Duplicate rows are: ",df_cleaned.duplicated().sum())
df_cleaned = df_cleaned.drop_duplicates()
df_cleaned.head(10)
df_cleaned['Gender'] = df_cleaned['Gender'].str.lower()
df_cleaned['Gender'] = df_cleaned['Gender'].replace({'f': 'Female', 'm': 'Male'})
df_cleaned.head(10)
```

No irrelevant columns in this case, but here's how you'd drop them ifneeded:
df_cleaned = df_cleaned.drop(['Irrelevant_Column'], axis=1)


```python
cleaned_file_path = 'cleaned_sample_dataset.csv'
df_cleaned.to_csv(cleaned_file_path, index=False)
```
